```python
# saving links by date
import datetime
import pandas as pd

def make_2_digits(num):
    if len(num) == 1:
        return "0" + num
    return num

valid_links = []

# Loop through dates from January 1 to December 31 2024
start_date = datetime.date(2001, 1, 1)
end_date = datetime.date(2001, 12, 31)
delta = datetime.timedelta(days=1)
while start_date <= end_date:
    link = "https://phillipian.net/" + str(start_date.year) + "/" + make_2_digits(str(start_date.month)) + "/" + make_2_digits(str(start_date.day)) + "/"
    # Check if link is valid (i.e. if it returns a 200 status code)
    page = requests.get(link)
    if page.status_code == 200:
        valid_links.append(link)
        # print(link)
    start_date += delta
```

```python
with open('valid_links.txt', 'w') as f:
    for line in valid_links:
        f.write(f"{line}\n")
```

```python
# takes a valid link and scrapes all articles from that date
from newspaper import Article
import requests
from bs4 import BeautifulSoup

def make_2_digits(num):
    if len(num) == 1:
        return "0" + num
    return num

f = open("date_valid_links_final.txt", "r")

for line in f.read().split("\n"):
    page = requests.get(line)
    soup = BeautifulSoup(page.content)
    
    links = soup.find_all("a", href=True)
    clean_links = []
    for link in links:
        if line in link['href']:
            # print(link['href'])
            if link['href'] not in clean_links:
                clean_links.append(link['href'])
    
    # print(clean_links)

    for link in clean_links:
        page = requests.get(link)
        soup1 = BeautifulSoup(page.content)

        article = Article(link)
        article.download()
        article.parse()
       
        # Get category
        category = soup1.find("div", {"class": "post-categories"})
        category = category.find("a").next.strip()
        # print(category)

        # Get date
        date = '-'.join(link.split("/")[3:6])
        # print(date)

        # Get the author of an article
        # author = article.author
        author = article.authors[:article.authors.index(".Wp-Block-Co-Authors-Plus-Coauthors.Is-Layout-Flow")]
        # print(article.authors[:article.authors.index(".Wp-Block-Co-Authors-Plus-Coauthors.Is-Layout-Flow")])

         # To print out the full text
        content = article.text
        # print(article.text)
        
        df = pd.concat([df, pd.DataFrame({'URL':link, 'Date':date, 'Category':category, 'Content':content, 'Author':author})])
        # break
    break

f.close()
```
